## Lemmetization 

Lemmatization is a technique in which the lemmetizer finds the root stem not the root word unlike the stemming techniques. <br>
NLTK->lemmetize->WordNetLemmetizer. <br>

Lemmetizer can be used to find the root stem with the following ways , by initializing pos -> VERB , ADJECTIVE , NOUN , ADVERB 

### USE CASES 
1. Q&A
2. Chatbots 
3. Text Summarization


### Lemmetization takes more time than stemming,
as it compares the words from a very big corpus and returns the output

```python
import nltk
nltk.download('wordnet')
```

### Pos 
1. Noun - n
2. verb - v 
3. adjective - a
4. adverb - r

```python
from nltk.stem import WordNetLemmatizer

lemmetizer = WordNetLemmatizer()

lemmetizer.lemmatize("going",pos='n')
```

```python
words = ['goes','going','gone','sporty','sporting','surprise','eating','eaten','eat','fairly','fair','forgotten']

for i in words:
    print(i + "----->" + lemmetizer.lemmatize(i,pos='v'))
```

### Thus Lemmetizer has multiple use cases than Stemming 
### but it is slow as it searches the words in the wordnet corpus. thus takes time but gives good output
