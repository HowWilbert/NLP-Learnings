## Lemmetization 

Lemmatization is a technique in which the lemmetizer finds the root stem not the root word unlike the stemming techniques. <br>
NLTK->lemmetize->WordNetLemmetizer. <br>

Lemmetizer can be used to find the root stem with the following ways , by initializing pos -> VERB , ADJECTIVE , NOUN , ADVERB 

### USE CASES 
1. Q&A
2. Chatbots 
3. Text Summarization


### Lemmetization takes more time than stemming,
as it compares the words from a very big corpus and returns the output

```python
import nltk
nltk.download('wordnet')
```

### Pos 
1. Noun - n
2. verb - v 
3. adjective - a
4. adverb - r

```python
from nltk.stem import WordNetLemmatizer

lemmetizer = WordNetLemmatizer()

lemmetizer.lemmatize("going",pos='n')
```

```python
words = ['goes','going','gone','sporty','sporting','surprise','eating','eaten','eat','fairly','fair','forgotten']

for i in words:
    print(i + "----->" + lemmetizer.lemmatize(i,pos='v'))
```

### Thus Lemmetizer has multiple use cases than Stemming 
### but it is slow as it searches the words in the wordnet corpus. thus takes time but gives good output

```python
from collections import deque
from typing import List, Tuple

def solve(
    rows: int,
    cols: int,
    grid: List[List[int]],
    start_row: int,
    start_col: int,
    end_row: int,
    end_col: int,
    gates,
    max_teleports: int,
) -> int:

    # Build teleport map
    teleport = {}

    for gate in gates:
        # Supports both formats:
        # (fr, fc, tr, tc)
        # ((fr, fc), (tr, tc))
        if len(gate) == 4:
            fr, fc, tr, tc = gate
        else:
            (fr, fc), (tr, tc) = gate

        teleport[(fr, fc)] = (tr, tc)

    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

    visited = [[[False] * (max_teleports + 1)
                for _ in range(cols)]
                for _ in range(rows)]

    q = deque()
    q.append((start_row, start_col, 0, 0))
    visited[start_row][start_col][0] = True

    while q:
        r, c, used, steps = q.popleft()

        if r == end_row and c == end_col:
            return steps

        # Move in four directions
        for dr, dc in directions:
            nr, nc = r + dr, c + dc

            if (0 <= nr < rows and
                0 <= nc < cols and
                grid[nr][nc] == 0 and
                not visited[nr][nc][used]):

                visited[nr][nc][used] = True
                q.append((nr, nc, used, steps + 1))

        # Use teleport if available
        if used < max_teleports and (r, c) in teleport:
            nr, nc = teleport[(r, c)]

            if (0 <= nr < rows and
                0 <= nc < cols and
                grid[nr][nc] == 0 and
                not visited[nr][nc][used + 1]):

                visited[nr][nc][used + 1] = True
                q.append((nr, nc, used + 1, steps + 1))

    return -1
```
