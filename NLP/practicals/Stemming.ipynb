### Stemming:
It is an Important process of text preprocessing, where we try to find the stem word from the set of similar words. 
example - goes, hoing , gone : [go is stem word]
        - eating , eaten , eat : [eat is the stem word]

##### But it is not always accurate. 
#### Some of the Popular Stemming techniques using nltk.stem are 
1. PorterStemmer 
2. RegexpStemmer 
3. SnowballStemmer 

## PorterStemmer

```python
import nltk
from nltk.stem import PorterStemmer

ps = PorterStemmer()
words = ['goes','going','gone','sporty','sporting','surprise','eating','eaten','eat','fairly','fair','forgotten']

for i in words:
    print(i+"---->"+ps.stem(i))
```

#### Observation 
1. many words after PorterStemming don't hold any actual meaning. 
2. thus this technique is not used for chatbot building

## RegexpStemmer

takes regular expressions as input and then removes thoes words from the strings it looks at <br>
takes input -> regular expression , min = ( number of minimum words that is required to apply regular expression )

```python
from nltk.stem import RegexpStemmer

rs = RegexpStemmer('ing$|e$|s$|able$|gly$',min = 4)

words = ['goes','going','gone','sporty','sporting','surprise','eating','eaten','eat','fairly','fair','forgotten']
for i in words:
    print(i + "------>"+ rs.stem(i))
```

### Observation 

1. Works only on thoes which satisfy the patterns given in the regular expression
2. we have to manually define the regular expressions.

3. good when we know which words are to come

## SnowballStemmer

By far the best Stemmer among the 3<br>
Provides the best result out of the 3<br>
still not that great still misses few words, but works better on few words than PorterStemmer

#### Works on Multiple Languages

```python
from nltk.stem import SnowballStemmer

sbs = SnowballStemmer('english')

words = ['goes','going','gone','sporty','sporting','surprise','eating','eaten','eat','fairly','fair','forgotten']

words = ['goes','going','gone','sporty','sporting','surprise','eating','eaten','eat','fairly','fair','forgotten']
for i in words:
    print(i + "------>"+ sbs.stem(i))
```
