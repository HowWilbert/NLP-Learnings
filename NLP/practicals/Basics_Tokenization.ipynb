```python
from nltk.tokenize import sent_tokenize,word_tokenize,wordpunct_tokenize, TreebankWordTokenizer
```

```python
with open('test.txt','r') as file:
    content = file.read()

print(content)
```

```python
import nltk
nltk.download('punkt_tab')
```

#### 1. Corpus to Documents

```python
sentences = sent_tokenize(content)
for sentence in sentences:
    print(sentence)
```

#### 2. Documents to Words

```python
for i in sentences:
    print(word_tokenize(i))
```

#### Observation about word tokenize:
1. notice the " 's " 

### 3. Corpus to Words

```python
words = word_tokenize(content)
for i in words:
    print(i)
```

### Usecase of WordPunct_tokenize

```python
wordpunct = wordpunct_tokenize(content)
wordpunct
```

#### Observation 
1. Note the " 's " is treated as " ' " and " s " separately

#### Use case of TreebankWordtokenizer

```python
treebank = TreebankWordTokenizer()
tree = treebank.tokenize(content)
for i in tree:
    print(i)
```

#### Observation
1. Only the last full stop is seprate,
2. The rest like Bire. are together only
